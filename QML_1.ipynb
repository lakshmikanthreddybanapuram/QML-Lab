# **TASK 1: Born Rule for Measurement Probabilities**

### **Aim**
To compute **measurement probabilities** of quantum states using the **Born rule**.



### **Procedure**

Quantum software development involves designing algorithms and applications that run on **quantum processors (Qubits)**.  
Several software development kits (SDKs) and platforms have emerged to support quantum programming, such as:

1. **Qiskit (IBM Quantum):** Python-based SDK supporting quantum circuit design, simulation, and real hardware execution.  
2. **Cirq (Google):** Focuses on near-term quantum algorithms and hardware integration.  
3. **PennyLane (Xanadu):** Integrates quantum computing with machine learning workflows.  
4. **TensorFlow Quantum:** Converts classical layers into quantum-based learning circuits.  
5. **Amazon Braket:** Provides cloud-based access to multiple quantum hardware providers.



#### **Steps Followed**

1. Define **quantum superposition states**.  
2. Apply the **Born rule** to compute measurement probabilities.  
3. Normalize the computed probabilities.  
4. Visualize results using **bar charts** for comparison between quantum states.

The **Born rule** states that the probability of measuring a state $|x\rangle$ from a quantum state $|\psi\rangle$ is given by:  
$P(x) = |\langle x | \psi \rangle|^2$

For example, for a superposition state $|\psi\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$,  
the measurement probabilities for $|0\rangle$ and $|1\rangle$ are both $\frac{1}{2}$.

```python
import numpy as np
import matplotlib.pyplot as plt

print("\n" + "="*50)
print("TASK 1: BORN RULE - MEASUREMENT PROBABILITIES")
print("="*50)

# Define Born rule function
def born_rule_probabilities(psi):
    probabilities = np.abs(psi)**2
    return probabilities / np.sum(probabilities)  # Normalize

# Create superposition states
psi_1 = np.array([1/np.sqrt(2), 1/np.sqrt(2)])   # |+⟩ state
psi_2 = np.array([1/np.sqrt(3), np.sqrt(2/3)])   # Custom superposition

# Display results
print("Superposition state 1: |ψ₁⟩ =", psi_1)
print("Measurement probabilities:", born_rule_probabilities(psi_1))
print("Superposition state 2: |ψ₂⟩ =", psi_2)
print("Measurement probabilities:", born_rule_probabilities(psi_2))

# Visualization
states = ['|0⟩', '|1⟩']
probs_1 = born_rule_probabilities(psi_1)
probs_2 = born_rule_probabilities(psi_2)

plt.figure(figsize=(10, 4))

# First state
plt.subplot(1, 2, 1)
plt.bar(states, probs_1, color='blue', alpha=0.7)
plt.title('State |ψ₁⟩ Probabilities')
plt.ylabel('Probability')
plt.ylim(0, 1)

# Second state
plt.subplot(1, 2, 2)
plt.bar(states, probs_2, color='red', alpha=0.7)
plt.title('State |ψ₂⟩ Probabilities')
plt.ylabel('Probability')
plt.ylim(0, 1)

plt.tight_layout()
plt.show()
```
